# Russell 3000

```python
r3000 = pd.read_csv("Russell3000EarningsAnnouncements.csv")["ticker"].tolist()
```

```python
url = "https://eodhd.com/api/eod/AAPL.US?period=d&api_token=6608304b3de2b5.23968015&fmt=csv&from=2019-10-01&to=2024-10-01"
data = requests.get(url).content
data_str = data.decode("utf-8")
df = pd.read_csv(StringIO(data_str))
print(df)
```

```python
data = {}
```

```python
i = 0
for ticker in r3000:
    print(f"{i}/{len(r3000)}")
    resp = api.get_eod_historical_stock_market_data(
        symbol=f"{ticker}.US",
        # from_date="2016-10-01",
        to_date="2024-10-01",
        period="d",
        order="a",
    )
    data[ticker] = pd.DataFrame(resp)
    i += 1
```

```python
with open("russell3000.pickle", "wb") as f:
    pickle.dump(data, f, protocol=pickle.HIGHEST_PROTOCOL)
```
